```python
## FFT

import numpy as np

def bitrevorder(x):
    n = len(x)
    n_bits = int(np.log2(n))

    # Perform bit reversal
    for i in range(n):
        j = int('{:0{width}b}'.format(i, width=n_bits)[::-1], 2)
        if j > i:
            x[i], x[j] = x[j], x[i]

    return x

def FFT(x):
    p = np.ceil(np.log2(len(x)))  # Next power of 2
    x = np.concatenate((x, np.zeros(2**int(p) - len(x))))  # Zero-padding
    N = len(x)
    S = int(np.log2(N))
    Half = N // 2
    
    for stage in range(1, S + 1):
        for index in range(0, N, N // (2 ** (stage - 1))):
            for n in range(Half):
                pos = n + index
                pow = (2 ** (stage - 1)) * n
                w = np.exp((-1j * 2 * np.pi * pow)/ N)
                a = x[pos] + x[pos + Half]
                b = (x[pos] - x[pos + Half]) * w
                x[pos] = a
                x[pos + Half] = b
        
        Half //= 2
    
    y = bitrevorder(x)
    return y

def iFFT(x):
        
    y = FFT(x)/x.size
    
    return np.concatenate(([y[0]], y[1:][::-1]))

# Test the function
x = np.array([1, 2, 3, 4, 5, 6, 7, 8], dtype = complex)
y = FFT(x)
x_ret = iFFT(y)
print(np.round(y)-np.round(np.fft.fft(x)))

print(np.round(x_ret))
```
