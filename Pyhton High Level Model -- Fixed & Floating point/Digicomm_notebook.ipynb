# Performance of Digital Modulations over Wireless Channels

```python
import numpy as np
import matplotlib, matplotlib.pyplot as plt

print('Matplotlib', matplotlib.__version__)
print('NumPy', np.__version__)
```

```python
from testing_script import *
```

```python
# Input parameters
nSym = 10**6 # Number of symbols to transmit
EbN0dBs = np.arange(start=-4,stop = 12, step = 2) # Eb/N0 range in dB for simulation
```

```python
def plot_res(EbN0dBs, SER_sim, SER_theo, mod_type, arrayOfM, plot_type): 

    colors = plt.cm.jet(np.linspace(0,1,len(arrayOfM))) # colormap
    fig, ax = plt.subplots(nrows=1,ncols = 1)

    for i, M in enumerate(arrayOfM):
        ax.semilogy(EbN0dBs,SER_sim[i,:],color = colors[i],marker='o',linestyle='',label='Sim '+str(M)+'-'+mod_type.upper())
        ax.semilogy(EbN0dBs,SER_theo[i,:],color = colors[i],linestyle='-',label='Theory'+str(M)+'-'+mod_type.upper())


    ax.set_xlabel('Eb/N0(dB)');ax.set_ylabel('SER ($P_s$)')
    ax.set_title('Probability of Symbol Error for M-'+str(mod_type)+' over ' + plot_type)
    ax.legend()
    fig.show()
```

```python
def plot_rician(EbN0dBs, SER_sim, SER_theo, mod_type, M, K_dBs): 

    colors = plt.cm.jet(np.linspace(0,1,len(K_dBs))) # colormap
    fig, ax = plt.subplots(nrows=1,ncols = 1)

    for i, K_dB in enumerate(K_dBs):
        ax.semilogy(EbN0dBs,SER_sim[i,:],color = colors[i],marker='o',linestyle='',label='Sim K='+str(K_dB)+' dB')
        ax.semilogy(EbN0dBs,SER_theo[i,:],color = colors[i],linestyle='-',label='TheoryK='+str(K_dB)+' dB')


    ax.set_xlabel('Eb/N0(dB)');ax.set_ylabel('SER ($P_s$)')
    ax.set_title('Probability of Symbol Error for ' + str(M) + '-' + str(mod_type)+' over Rician flat fading channel')
    ax.legend()
    fig.show()
```
